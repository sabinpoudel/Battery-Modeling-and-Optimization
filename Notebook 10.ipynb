```python
# ======================================================================================
# STEP 10 — OPTIONAL-IMPORT FUNCTION
# ======================================================================================

OPTIONAL_IMPORT_ERRORS: dict[str, str] = {}


def optional_import(
    module_name: str,
    alias: Optional[str] = None,
):
    """
    Import an optional module.

    Returns
    -------
    module
        Imported module when successful.

    None
        Returned when the package is unavailable or cannot be imported.
    """
    try:
        module = importlib.import_module(module_name)

        if alias is not None:
            globals()[alias] = module

        return module

    except Exception as exc:
        OPTIONAL_IMPORT_ERRORS[module_name] = (
            f"{type(exc).__name__}: {exc}"
        )

        return None


print("Optional-import function created.")
```

```python
# ======================================================================================
# STEP 17 — GENERAL UTILITY FUNCTIONS
# ======================================================================================

def natural_sort_key(value: Any) -> list[Any]:
    """
    Produce a natural sorting key.

    Example
    -------
    S2 appears before S10.
    """
    return [
        int(token) if token.isdigit() else token.lower()
        for token in re.split(
            r"(\d+)",
            str(value),
        )
    ]


def ensure_directory(
    path: Path | str,
) -> Path:
    """
    Create a directory when it does not exist.
    """
    resolved_path = (
        Path(path)
        .expanduser()
        .resolve()
    )

    resolved_path.mkdir(
        parents=True,
        exist_ok=True,
    )

    return resolved_path


def sha256_file(
    path: Path | str,
    chunk_size: int = 1024 * 1024,
) -> str:
    """
    Calculate the SHA-256 hash of a file.
    """
    digest = hashlib.sha256()

    with Path(path).open("rb") as stream:
        for block in iter(
            lambda: stream.read(chunk_size),
            b"",
        ):
            digest.update(block)

    return digest.hexdigest()


def json_default(value: Any):
    """
    Convert common scientific Python objects into JSON-compatible forms.
    """
    if isinstance(value, Path):
        return str(value)

    if isinstance(value, np.integer):
        return int(value)

    if isinstance(value, np.floating):
        return float(value)

    if isinstance(value, np.ndarray):
        return value.tolist()

    if isinstance(value, pd.Timestamp):
        return value.isoformat()

    if isinstance(value, datetime):
        return value.isoformat()

    try:
        if pd.isna(value):
            return None
    except Exception:
        pass

    raise TypeError(
        f"Object of type {type(value).__name__} "
        "is not JSON serializable."
    )


def save_json(
    data: Any,
    path: Path | str,
    indent: int = 2,
) -> Path:
    """
    Save a Python object as a UTF-8 JSON file.
    """
    output_path = (
        Path(path)
        .expanduser()
        .resolve()
    )

    output_path.parent.mkdir(
        parents=True,
        exist_ok=True,
    )

    with output_path.open(
        "w",
        encoding="utf-8",
    ) as stream:
        json.dump(
            data,
            stream,
            indent=indent,
            ensure_ascii=False,
            default=json_default,
        )

    return output_path


def dataframe_memory_mb(
    frame: pd.DataFrame,
) -> float:
    """
    Return the deep memory usage of a DataFrame in MiB.
    """
    return float(
        frame.memory_usage(
            index=True,
            deep=True,
        ).sum()
        / (1024**2)
    )


def root_mean_squared_error(
    y_true: Sequence[float],
    y_pred: Sequence[float],
) -> float:
    """
    Calculate RMSE while ignoring paired non-finite observations.
    """
    true_array = np.asarray(
        y_true,
        dtype=float,
    )

    predicted_array = np.asarray(
        y_pred,
        dtype=float,
    )

    finite_mask = (
        np.isfinite(true_array)
        & np.isfinite(predicted_array)
    )

    if not finite_mask.any():
        return np.nan

    squared_error = np.square(
        true_array[finite_mask]
        - predicted_array[finite_mask]
    )

    return float(
        np.sqrt(
            np.mean(squared_error)
        )
    )


@contextmanager
def execution_timer(
    label: str = "Operation",
):
    """
    Measure and print the execution time of a code block.
    """
    start_time = time.perf_counter()

    try:
        yield

    finally:
        elapsed_time = (
            time.perf_counter()
            - start_time
        )

        print(
            f"{label} completed in "
            f"{elapsed_time:,.3f} seconds."
        )


print("Reusable utility functions created.")
```

```python
# ======================================================================================
# STEP 18 — PROJECT-ROOT DETECTION
# ======================================================================================

def find_project_root(
    candidates: Sequence[Path],
) -> Path:
    """
    Select the most likely battery-project root.
    """
    scored_candidates = []

    expanded_candidates = list(candidates)

    expanded_candidates.extend(
        [
            Path.cwd(),
            *Path.cwd().parents,
        ]
    )

    for candidate in expanded_candidates:
        candidate = (
            candidate
            .expanduser()
            .resolve()
        )

        if not candidate.exists():
            continue

        if not candidate.is_dir():
            continue

        score = 0

        expected_subdirectories = [
            "data",
            "notebooks",
            "src",
            "figures",
            "results",
            "models",
        ]

        for name in expected_subdirectories:
            if (candidate / name).exists():
                score += 1

        if "batter" in candidate.name.lower():
            score += 3

        scored_candidates.append(
            {
                "score": score,
                "path": candidate,
            }
        )

    if not scored_candidates:
        raise FileNotFoundError(
            "No valid battery-project root was found."
        )

    scored_candidates = sorted(
        scored_candidates,
        key=lambda item: (
            -item["score"],
            len(str(item["path"])),
        ),
    )

    return scored_candidates[0]["path"]


PROJECT_ROOT = find_project_root(
    PROJECT_ROOT_CANDIDATES
)

print("Detected project root:")
print(PROJECT_ROOT)
```

```python
# ======================================================================================
# STEP 19 — PUBLISHED-DIRECTORY DETECTION
# ======================================================================================

def find_published_directories(
    project_root: Path,
) -> list[dict[str, Any]]:
    """
    Locate directories containing published Parquet files.
    """
    preferred_candidates = [
        project_root / "data" / "published",
        project_root / "data" / "processed" / "published",
        project_root / "data" / "processed" / "publication",
        project_root / "published",
        project_root / "publication",
    ]

    candidate_directories = []

    for candidate in preferred_candidates:
        if candidate.exists() and candidate.is_dir():
            candidate_directories.append(
                candidate.resolve()
            )

    for candidate in project_root.rglob("*"):
        if not candidate.is_dir():
            continue

        if candidate.name.lower() in PUBLISHED_DIRECTORY_NAMES:
            candidate_directories.append(
                candidate.resolve()
            )

    candidate_directories = sorted(
        set(candidate_directories),
        key=lambda path: natural_sort_key(str(path)),
    )

    results = []

    for candidate in candidate_directories:
        parquet_files = sorted(
            [
                path
                for path in candidate.rglob("*")
                if path.is_file()
                and path.suffix.lower() in PARQUET_SUFFIXES
            ],
            key=lambda path: natural_sort_key(str(path)),
        )

        if not parquet_files:
            continue

        total_size_bytes = sum(
            path.stat().st_size
            for path in parquet_files
        )

        results.append(
            {
                "path": candidate,
                "parquet_count": len(parquet_files),
                "total_parquet_bytes": total_size_bytes,
                "total_parquet_mb": (
                    total_size_bytes / (1024**2)
                ),
            }
        )

    results.sort(
        key=lambda item: (
            -item["parquet_count"],
            -item["total_parquet_bytes"],
            len(str(item["path"])),
        )
    )

    return results


PUBLISHED_DIRECTORY_CANDIDATES = (
    find_published_directories(
        PROJECT_ROOT
    )
)

if not PUBLISHED_DIRECTORY_CANDIDATES:
    raise FileNotFoundError(
        "No published directory containing Parquet files was found.\n"
        "Run Notebook 9 first or assign the correct published path."
    )


published_candidate_df = pd.DataFrame(
    PUBLISHED_DIRECTORY_CANDIDATES
)

display(published_candidate_df)


PUBLISHED_DIR = (
    PUBLISHED_DIRECTORY_CANDIDATES[0]["path"]
)

print("Selected published directory:")
print(PUBLISHED_DIR)
```

```python
# ======================================================================================
# STEP 22 — DEPENDENCY REPORT
# ======================================================================================

# Build one dependency record for every package registered in STEP 3.
dependency_records = []

for package in PACKAGE_REGISTRY:
    module_name = package["module"]
    distribution_name = package["distribution"]

    # ------------------------------------------------------------------
    # Check whether the importable Python module is available.
    # ------------------------------------------------------------------
    try:
        module_available = (
            importlib.util.find_spec(module_name)
            is not None
        )
    except Exception:
        module_available = False

    # ------------------------------------------------------------------
    # Obtain the installed distribution version.
    # The distribution name may differ from the module name:
    # sklearn -> scikit-learn, yaml -> PyYAML, etc.
    # ------------------------------------------------------------------
    try:
        package_version = importlib.metadata.version(
            distribution_name
        )
    except importlib.metadata.PackageNotFoundError:
        package_version = None
    except Exception:
        package_version = None

    record = {
        "module": module_name,
        "distribution": distribution_name,
        "group": package["group"],
        "required": bool(package["required"]),
        "available": bool(module_available),
        "version": package_version,
        "import_status": None,
        "import_error": None,
    }

    # ------------------------------------------------------------------
    # Determine the actual import status.
    # ------------------------------------------------------------------
    if module_name in OPTIONAL_IMPORT_ERRORS:
        record["import_status"] = "FAILED"
        record["import_error"] = (
            OPTIONAL_IMPORT_ERRORS[module_name]
        )

    elif module_name in sys.modules:
        record["import_status"] = "IMPORTED"
        record["import_error"] = None

    elif module_available:
        try:
            importlib.import_module(module_name)

            record["import_status"] = "IMPORTED"
            record["import_error"] = None

        except Exception as exc:
            record["import_status"] = "FAILED"
            record["import_error"] = (
                f"{type(exc).__name__}: {exc}"
            )

    else:
        record["import_status"] = "MISSING"
        record["import_error"] = (
            "Package is not installed."
        )

    dependency_records.append(record)


# ======================================================================================
# CREATE THE DEPENDENCY-REPORT DATAFRAME
# ======================================================================================

dependency_report_df = pd.DataFrame(
    dependency_records
)

dependency_report_df = dependency_report_df.sort_values(
    by=[
        "required",
        "group",
        "module",
    ],
    ascending=[
        False,
        True,
        True,
    ],
    kind="mergesort",
    ignore_index=True,
)

display(dependency_report_df)


# ======================================================================================
# IDENTIFY MISSING REQUIRED AND OPTIONAL PACKAGES
# ======================================================================================

missing_required_packages = (
    dependency_report_df.loc[
        dependency_report_df["required"]
        & (
            dependency_report_df["import_status"]
            == "MISSING"
        ),
        "module",
    ]
    .astype(str)
    .tolist()
)

missing_optional_packages = (
    dependency_report_df.loc[
        (~dependency_report_df["required"])
        & (
            dependency_report_df["import_status"]
            == "MISSING"
        ),
        "module",
    ]
    .astype(str)
    .tolist()
)


# Required packages that are either missing or failed to import.
failed_required_imports = dependency_report_df.loc[
    dependency_report_df["required"]
    & (
        dependency_report_df["import_status"]
        != "IMPORTED"
    )
].copy()


if not failed_required_imports.empty:
    print(
        "\nRequired packages that could not be imported:"
    )

    display(failed_required_imports)

    if STRICT_REQUIRED_DEPENDENCIES:
        raise ImportError(
            "One or more required packages are missing "
            "or could not be imported."
        )


# Optional package problems are reported but do not normally stop execution.
failed_optional_imports = dependency_report_df.loc[
    (~dependency_report_df["required"])
    & (
        dependency_report_df["import_status"]
        == "FAILED"
    )
].copy()


if not failed_optional_imports.empty:
    print(
        "\nOptional packages that failed during import:"
    )

    display(failed_optional_imports)

    if STRICT_OPTIONAL_DEPENDENCIES:
        raise ImportError(
            "One or more optional packages could not be imported."
        )


print(
    "\nDependency-report summary"
)

print(
    "Registered packages      :",
    len(dependency_records),
)

print(
    "Successfully imported    :",
    int(
        (
            dependency_report_df["import_status"]
            == "IMPORTED"
        ).sum()
    ),
)

print(
    "Missing required packages:",
    len(missing_required_packages),
)

print(
    "Missing optional packages:",
    len(missing_optional_packages),
)

print(
    "Failed package imports   :",
    int(
        (
            dependency_report_df["import_status"]
            == "FAILED"
        ).sum()
    ),
)
```
